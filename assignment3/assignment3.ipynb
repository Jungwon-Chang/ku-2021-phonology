## F0(Hz) plotting

```python
import matplotlib.pyplot as plt
import pandas as pd
df_pitch = pd.read_csv('./yourgoodpants.csv', sep=',')
df_pitch.head(20)
```

```python
df_pitch.loc[df_pitch.Time_s == "--undefined--", "Time_s"] = None
df_pitch.loc[df_pitch.F0_Hz == "--undefined--", "F0_Hz"] = None
df_pitch.head(10)
```

```python
time = df_pitch.Time_s
f0 = df_pitch.F0_Hz

plt.plot(time, f0, 'ro', label='F0 (Hz)')
plt.xlabel('Time(s)')
plt.ylabel('F0')
plt.legend(loc='best', ncol=2, fontsize=14, frameon=True, shadow=True)
plt.show()
```

## intensity(dB) plotting

```python
df_intensity = pd.read_csv('./yourgoodpants_intensity.csv', sep=',')
df_intensity.head(20)
```

## Formant plotting

```python
df_formants = pd.read_csv('./yourgoodpants_formants.csv', sep=',')
df_formants.head(20)
```

```python
a = df_formants.F3_Hz == "--undefined--", "F3_Hz"
```

```python
df_formants.loc[df_formants.Time_s == "--undefined--", "Time_s"] = None
df_formants.loc[df_formants.F1_Hz == "--undefined--", "F1_Hz"] = None
df_formants.loc[df_formants.F2_Hz == "--undefined--", "F2_Hz"] = None
df_formants.loc[df_formants.F3_Hz == "--undefined--", "F3_Hz"] = None
df_formants.loc[df_formants.F4_Hz == "--undefined--", "F4_Hz"] = None
df_formants.head(20)
```

```python
time = df_formants.Time_s
f1 = df_formants.F1_Hz
f2 = df_formants.F2_Hz
f3 = df_formants.F3_Hz.astype(float)
f4 = df_formants.F4_Hz.astype(float)
```

```python
plt.plot(time, f1, 'ro', label='F1 (Hz)')
plt.plot(time, f2, 'bo', label='F2 (Hz)')
plt.plot(time, f3, 'go', label='F3 (Hz)')
plt.plot(time, f4, 'yo', label='F4 (Hz)')
plt.xlabel('Time(s)')
plt.ylabel('Hz')
plt.legend(loc='best', ncol=2, fontsize=14, frameon=True, shadow=True)
plt.show()
```
