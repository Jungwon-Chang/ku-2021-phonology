```python
import numpy as np
import IPython.display as ipd
```

### Play musical notes

```python
# basic functions
def phasor_sine(amp, sr, dur, freq, phase):
    t = np.arange(1, sr * dur+1)/sr
    theta = t * 2*np.pi * freq + phase
    s = amp * np.sin(theta)
    return s, t, theta

def getFrequency(note):
    A4=440
    notes = ['A', 'A#', 'B', 'C', 'C#', 'D', 'D#', 'E', 'F', 'F#', 'G', 'G#']

    octave = int(note[2]) if len(note) == 3 else int(note[1])

    keyNumber = notes.index(note[0:-1]);
    
    if (keyNumber < 3) :
        keyNumber = keyNumber + 12 + ((octave - 1) * 12) + 1; 
    else:
        keyNumber = keyNumber + ((octave - 1) * 12) + 1; 

    return A4 * 2** ((keyNumber- 49) / 12)
```

```python
# set variables
amp = 1; sr = 10000; phase = 0

C4  = getFrequency('C4')
D4  = getFrequency('D4')
E4  = getFrequency('E4')
F4  = getFrequency('F4')
G4  = getFrequency('G4')
A4  = getFrequency('A4')
Bb4 = getFrequency('A#4')
B4  = getFrequency('B4')
C5  = getFrequency('C5')
D5  = getFrequency('D5')
E5  = getFrequency('E5')
F5  = getFrequency('F5')
B5  = getFrequency('B5')
C6  = getFrequency('C6')
```

```python
# The Coffin Dance
notes = [A4, 0,  A4, E5, D5, C5, B4, 0, B4, 0, B4, D5, C5, B4, A4, 0, A4, C6,
         B5, C6, B5, C6, A4, 0, A4, C6, B5, C6, B5, C6]

durs  = [0.5, 0.015, 0.235, 0.25, 0.5, 0.5, 0.5, 0.015, 0.235, 0.015, 0.235, 0.5,
         0.25, 0.25, 0.5, 0.015, 0.235, 0.25, 0.25, 0.25, 0.25, 0.25, 0.5, 0.015,
         0.235, 0.25, 0.25, 0.25, 0.25, 0.25]
```

```python
s_all = np.empty(0)

for note, dur in zip(notes, durs):
    s, t, th= phasor_sine(amp, sr, dur, note, phase)
    s_all = np.concatenate((s_all, s))

ipd.Audio(s_all, rate=sr)
```
